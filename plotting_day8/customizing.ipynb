```python
import pandas as pd

url = "https://raw.githubusercontent.com/plotly/datasets/master/gapminder_with_codes.csv"
data = pd.read_csv(url)

data_2007 = data[data["year"] == 2007]

data_2007.head()
```

```python
import matplotlib.pyplot as plt

plt.style.use('ggplot')

fig, ax = plt.subplots()

ax.scatter(x=data_2007["gdpPercap"], y=data_2007["lifeExp"], alpha=0.5)

ax.set_xlabel("GDP (USD) per capita")
ax.set_ylabel("life expectancy (years)")
```

```python
fig, ax = plt.subplots()

ax.scatter(x=data_2007["gdpPercap"], y=data_2007["lifeExp"], alpha=0.5)

ax.set_xlabel("GDP (USD) per capita")
ax.set_ylabel("life expectancy (years)")

ax.set_xscale('log')
```

```python
fig, (ax_left, ax_right) = plt.subplots(1, 2, figsize=(7.0, 3.0), layout='constrained')

ax_left.scatter(x=data_2007["gdpPercap"], y=data_2007["lifeExp"], alpha=0.5)

x_label = "GDP (USD) per capita"

ax_left.set_xlabel(x_label)
ax_left.set_ylabel("life expectancy (years)")

ax_right.scatter(x=data_2007["gdpPercap"], y=data_2007["lifeExp"], alpha=0.5)
ax_right.set_xlabel(x_label)
ax_right.set_ylabel("life expectancy (years)")
ax_right.set_xscale('log')
```
