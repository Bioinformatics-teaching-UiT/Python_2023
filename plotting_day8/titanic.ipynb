```python
import pandas as pd

url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv"
data = pd.read_csv(url)

data.describe()
```

```python
df = data.loc[:, ['sex', 'survived']]
df = df.pivot_table(index='sex', columns=['survived'], aggfunc=len)

df.plot(kind='bar', stacked=True, ylabel='Counts', xlabel='Gender',
       title='Survival status by gender')
```

```python
df = data.loc[:, ['pclass', 'survived']]
df = df.pivot_table(index='pclass', columns=['survived'], aggfunc=len)

df.plot(kind='bar', stacked=True, ylabel='Counts', xlabel='Class',
       title='Survival status by class')
```
